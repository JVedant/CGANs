```python
import pandas as pd
import requests
import shutil
```

```python
df = pd.read_csv("../DATA/unsplash/selected_features.csv")
```

```python
df
```

```python
photo_description_na_ids = df[df.photo_description.isna()]['photo_id'].values
```

```python
ai_description_na_ids = df[df.ai_description.isna()]['photo_id'].values
```

```python
ai_description_na_ids
```

```python
len(ai_description_na_ids)
```

```python
no_description = set(photo_description_na_ids).intersection(set(ai_description_na_ids))
```

```python
no_description = list(no_description)
```

```python
df
```

```python
for each in df.iterrows():
    if each[1].photo_id in no_description:
        #print(each[0])
        df.drop(each[0], inplace=True)
```

```python
df.ai_description.isna().sum()
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df
```

```python
df.photo_description.fillna("NO DESCRIPTION", inplace=True)
df.ai_description.fillna("NO DESCRIPTION", inplace=True)
```

```python
error_counter = 0

## ref: https://towardsdatascience.com/how-to-download-an-image-using-python-38a75cfa21c
def get_data(url, photo_name):
    try:
        r = requests.get(url, stream=True)

        if r.status_code == 200:
        # Set decode_content value to True, otherwise the downloaded image file's size will be zero.
            r.raw.decode_content = True
            
            # Open a local file with wb ( write binary ) permission.
            with open((photo_name + ".png"),'wb') as f:
                shutil.copyfileobj(r.raw, f)
        else:
            error_counter += 1
    except:
        pass
```

```python
get_data(df.iloc[0].photo_image_url, df.iloc[0].photo_id)
```

```python
df.to_csv("../DATA/unsplash/selected_features_cleaned.csv", index=False)
```
